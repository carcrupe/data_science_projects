```python
A=2.5
B1=20
B2=15
n_elements=100

x1=np.linspace(-2.2,2.2,n_elements)
y1 = A*x1+B1
plt.plot(x1,y1, 'b--')

x2=np.random.randn(n_elements)
y2 = A*x2+B2
plt.plot(x2,y2, 'g.');
plt.xlim([-2.2,2.2])
```

```python
def createSigmoid(n_points):
    x = np.linspace(-20,20,n_points)
    y = (np.exp(x))/(np.exp(x)+1)
    plt.plot(x,y,'g.')

createSigmoid(100)
```
