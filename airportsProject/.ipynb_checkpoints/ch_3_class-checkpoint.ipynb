### For the whole dataframe with chunks

```python
s = pd.read_csv('searches.csv.bz2',sep="^", usecols=['Date','Destination'], chunksize=500000)

all_chunks = pd.DataFrame()
for i,sChunk in enumerate(s):
    
    sChunk=sChunk.dropna()
    filtered = sChunk[sChunk['Destination'].isin(['MAD','BCN','MLA'])]
    filtered['Plot_Date']=filtered['Date'].str[0:7]
    s_gr = filtered.groupby(['Plot_Date','Destination']).count()
    s_gr = s_gr.reset_index()
    all_chunks=all_chunks.append(s_gr)

all_chunks_gr=all_chunks.groupby(['Plot_Date','Destination']).sum().reset_index()
to_plot=all_chunks_gr.pivot('Plot_Date', 'Destination', 'Date')
to_plot.plot()
```
