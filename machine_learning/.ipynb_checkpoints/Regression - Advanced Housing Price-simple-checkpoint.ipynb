# House Pricing

## Problem Statement

We would like to predict the price house of a house based on its characteristics.

```python
%pylab inline
import pandas as pd
```

```python
df = pd.read_csv("house_prices.csv")
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
df=df[df["TotalSF"]<6000]
```

```python
# Data Preparation
X = df[['TotalSF']] # pandas DataFrame
y = df["SalePrice"] # pandas Series
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X,y)
lr.predict([[1000], [2000],[3000]])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1)
```

```python
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
lr.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
# Load the library
from sklearn.tree import DecisionTreeRegressor
# Create an instance
regd = DecisionTreeRegressor(max_depth=3)
# Fit the data
regd.fit(X,y)
```

```python
from sklearn.metrics import make_scorer

def corr(pred,y_test):
    return np.corrcoef(pred,y_test)[0][1]

# Put the scorer in cross_val_score
cross_val_score(regd,X,y,cv=5,scoring=make_scorer(corr))
```

```python
def bias(pred,y_test):
    return np.mean(pred-y_test)
# Put the scorer in cross_val_score
cross_val_score(regd,X,y,cv=5,scoring=make_scorer(bias))
```
